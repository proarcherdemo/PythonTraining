Histogram : data visualization used to display the frequency (or count) of data points that fall within specified ranges, called bins.

> Bins (or intervals):
Bins are the divisions or intervals that the data is grouped into. 
The range of the data is divided into equally sized intervals or bins, and each bin represents a range of values.

> Frequency:
The height of each bar in the histogram represents the frequency or the number of data points that fall into the respective bin.

> X-axis:
The x-axis represents the data values (such as the ranges or bins).

> Y-axis:
It shows how many observations (or data points) fall into each bin.

Scatter plot :  
Uses dots to represent the values of two variables. 
Shows the relationship between two variables, especially in terms of correlation or patterns. 

Points to Note :

> Axes:
The x-axis represents one variable (usually the independent variable).
The y-axis represents the other variable (usually the dependent variable).

>Data Points:
The coordinates of each point are determined by the values of the two variables it represents.

>Correlation:

* Positive correlation: If the points tend to rise together (move up to the right), the variables are positively correlated.
    
* Negative correlation: If the points tend to move in opposite directions (one goes up as the other goes down), the variables are negatively correlated.
    
* No correlation: If the points appear scattered without any clear pattern, the variables have little or no correlation.

>Outliers:
Scatter plots can also help identify outliers, which are data points that significantly differ from other points in the dataset.

A box plot (also known as a box-and-whisker plot) is a standardized way of displaying the distribution of data based on a five-number summary:

* Minimum
* First quartile (Q1)  25th percentile
* Median (Q2)  50th percentile
* Third quartile (Q3)  75th percentile
* Maximum

Box plots are particularly useful for visualizing the spread and skewness of data, as well as identifying outliers.


>Box:
The box represents the interquartile range (IQR), which is the range between the first (Q1) and third quartile (Q3). The box is drawn from Q1 to Q3, and the width of the box shows the spread of the middle 50% of the data.

>Median:
A line inside the box marks the median (Q2), which divides the data into two equal halves.

>Whiskers:
The whiskers are the lines extending from the box to the minimum and maximum values in the data. Whiskers usually extend up to 1.5 times the interquartile range (IQR) from Q1 and Q3. Any data points outside this range are considered outliers.

>Outliers:
Outliers are any points that fall outside of the whiskers. 

Interpretation:
* Spread of Data: The width of the box represents the spread of the middle 50% of the data (the IQR). A wider box means greater variability.
* Skewness: If the median line is not centered in the box, the data may be skewed.
    * If the median is closer to Q1, the data is right-skewed (long tail on the right). 
    * If the median is closer to Q3, the data is left-skewed (long tail on the left).
* Outliers: Points outside the whiskers are considered outliers, which are unusually large or small values compared to the rest of the data.

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm

# Sample data: Heights in cm
heights = [160, 165, 170, 172, 168, 174, 169, 171, 167, 166]

# Calculate mean and standard deviation
mean = np.mean(heights)
std_dev = np.std(heights)

print(f"Mean: {mean:.2f} cm")
print(f"Standard Deviation: {std_dev:.2f} cm")

# Plotting
plt.figure(figsize=(8, 5))

# Histogram of the data
count, bins, ignored = plt.hist(heights, bins=8, density=True, alpha=0.6, color='skyblue', edgecolor='black')

# Plot normal distribution curve
xmin, xmax = plt.xlim()
x = np.linspace(xmin, xmax, 100)
p = norm.pdf(x, mean, std_dev)
plt.plot(x, p, 'r', linewidth=2)

plt.title('Heights of Students - Normal Distribution')
plt.xlabel('Height (cm)')
plt.ylabel('Probability Density')
plt.grid(True)
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm

# Data
heights = [160, 165, 170, 172, 168, 174, 169, 171, 167, 166]
mean = np.mean(heights)
std_dev = np.std(heights)

# x-values range
x = np.linspace(mean - 4*std_dev, mean + 4*std_dev, 500)
y = norm.pdf(x, mean, std_dev)

# Plot the normal curve
plt.figure(figsize=(10, 6))
plt.plot(x, y, color='black', label='Normal Distribution')

# Fill regions based on Empirical Rule
colors = ['#b3cde0', '#6497b1', '#005b96']
labels = ['68% (  1)', '95% (  2)', '99.7% (  3)']

for i in range(1, 4):
    x_fill = np.linspace(mean - i*std_dev, mean + i*std_dev, 500)
    y_fill = norm.pdf(x_fill, mean, std_dev)
    plt.fill_between(x_fill, y_fill, color=colors[i-1], alpha=0.6, label=labels[i-1])

# Mean line
plt.axvline(mean, color='red', linestyle='--', label=f'Mean () = {mean:.2f} cm')

# Labels and title
plt.title('Empirical Rule on Normal Distribution of Student Heights')
plt.xlabel('Height (cm)')
plt.ylabel('Probability Density')
plt.legend()
plt.grid(True)
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import expon

# Set random seed for reproducibility
np.random.seed(42)

# Parameters
lambda_rate = 1 / 6  # average of 6 minutes between events
n_samples = 20

# Generate exponential data
data = np.random.exponential(scale=1/lambda_rate, size=n_samples)

# Plot histogram
plt.figure(figsize=(10, 6))
count, bins, ignored = plt.hist(data, bins=10, density=True, alpha=0.6, color='skyblue', edgecolor='black', label='Observed data')

# Plot the exponential PDF
x = np.linspace(0, max(data) + 5, 1000)
pdf = lambda_rate * np.exp(-lambda_rate * x)
plt.plot(x, pdf, 'r-', lw=2, label=f'Exponential PDF\n($\\lambda$ = {lambda_rate:.3f})')

# Labels and legend
plt.title('Exponential Distribution: Time Between Calls')
plt.xlabel('Time Between Events (minutes)')
plt.ylabel('Density')
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
import numpy as np
import pandas as pd
import statsmodels.api as sm

# Simulate data
np.random.seed(0)
n = 100
hours_studied = np.random.normal(5, 2, n)
attendance_rate = np.random.normal(75, 10, n)
favorite_color = np.random.choice([0, 1, 2], size=n)  # Random, irrelevant variable

# Exam score is influenced by study and attendance, not color
exam_score = 10 + 5 * hours_studied + 0.5 * attendance_rate + np.random.normal(0, 5, n)

# Create DataFrame
df = pd.DataFrame({
    'HoursStudied': hours_studied,
    'AttendanceRate': attendance_rate,
    'FavoriteColor': favorite_color,
    'ExamScore': exam_score
})

# Total variance in exam scores
mean_exam_score = np.mean(exam_score)
ss_total = np.sum((exam_score - mean_exam_score) ** 2)

# Model 1
X1 = sm.add_constant(df[['HoursStudied']])
model1 = sm.OLS(exam_score, X1).fit()
ss_resid1 = np.sum(model1.resid ** 2)

# Model 2
X2 = sm.add_constant(df[['HoursStudied', 'AttendanceRate']])
model2 = sm.OLS(exam_score, X2).fit()
ss_resid2 = np.sum(model2.resid ** 2)

# Model 3
X3 = sm.add_constant(df[['HoursStudied', 'AttendanceRate', 'FavoriteColor']])
model3 = sm.OLS(exam_score, X3).fit()
ss_resid3 = np.sum(model3.resid ** 2)

# Compile and print results
results = pd.DataFrame({
    'Model': ['Model 1', 'Model 2', 'Model 3'],
    'Predictors': ['HoursStudied', 'HoursStudied + AttendanceRate', 'HoursStudied + AttendanceRate + FavoriteColor'],
    'SS_total': [ss_total]*3,
    'SS_residual': [ss_resid1, ss_resid2, ss_resid3],
    'R_squared': [model1.rsquared, model2.rsquared, model3.rsquared],
    'Adjusted_R_squared': [model1.rsquared_adj, model2.rsquared_adj, model3.rsquared_adj]
})

print(results.round(2))
```
