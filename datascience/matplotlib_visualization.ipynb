```python
# showing the legends 

# upper left, upper right, lower left, lower right
# upper center, lower center, center left, center right
# center, best

x = np.arange(10)
plt.plot(x, x)
plt.plot(x, 2*x)
plt.plot(x, 3*x)
plt.plot(x, 4*x)
plt.title("Plotting a line chart")
plt.xlabel("X-values")
plt.ylabel("Y-values")
plt.legend(['y = x', 'y = 2x', 'y = 3x', 'y = 4x'], loc='best') # default 'upper left'
plt.show() # to show the graph
```

Histogram : data visualization used to display the frequency (or count) of data points that fall within specified ranges, called bins.

> Bins (or intervals):
Bins are the divisions or intervals that the data is grouped into. 
The range of the data is divided into equally sized intervals or bins, and each bin represents a range of values.

> Frequency:
The height of each bar in the histogram represents the frequency or the number of data points that fall into the respective bin.

> X-axis:
The x-axis represents the data values (such as the ranges or bins).

> Y-axis:
It shows how many observations (or data points) fall into each bin.

Scatter plot :  
Uses dots to represent the values of two variables. 
Shows the relationship between two variables, especially in terms of correlation or patterns. 

Points to Note :

> Axes:
The x-axis represents one variable (usually the independent variable).
The y-axis represents the other variable (usually the dependent variable).

>Data Points:
The coordinates of each point are determined by the values of the two variables it represents.

>Correlation:

* Positive correlation: If the points tend to rise together (move up to the right), the variables are positively correlated.
    
* Negative correlation: If the points tend to move in opposite directions (one goes up as the other goes down), the variables are negatively correlated.
    
* No correlation: If the points appear scattered without any clear pattern, the variables have little or no correlation.

>Outliers:
Scatter plots can also help identify outliers, which are data points that significantly differ from other points in the dataset.

A box plot (also known as a box-and-whisker plot) is a standardized way of displaying the distribution of data based on a five-number summary:

* Minimum
* First quartile (Q1)  25th percentile
* Median (Q2)  50th percentile
* Third quartile (Q3)  75th percentile
* Maximum

Box plots are particularly useful for visualizing the spread and skewness of data, as well as identifying outliers.


>Box:
The box represents the interquartile range (IQR), which is the range between the first (Q1) and third quartile (Q3). The box is drawn from Q1 to Q3, and the width of the box shows the spread of the middle 50% of the data.

>Median:
A line inside the box marks the median (Q2), which divides the data into two equal halves.

>Whiskers:
The whiskers are the lines extending from the box to the minimum and maximum values in the data. Whiskers usually extend up to 1.5 times the interquartile range (IQR) from Q1 and Q3. Any data points outside this range are considered outliers.

>Outliers:
Outliers are any points that fall outside of the whiskers. 

Interpretation:
* Spread of Data: The width of the box represents the spread of the middle 50% of the data (the IQR). A wider box means greater variability.
* Skewness: If the median line is not centered in the box, the data may be skewed.
    * If the median is closer to Q1, the data is right-skewed (long tail on the right). 
    * If the median is closer to Q3, the data is left-skewed (long tail on the left).
* Outliers: Points outside the whiskers are considered outliers, which are unusually large or small values compared to the rest of the data.

```python
data = [data.CurrentSalary]

plt.boxplot(data)
```
