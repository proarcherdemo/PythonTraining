```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv("Salary_Data.csv")
print(data)
```

```python
data.info()
```

```python
plt.scatter(data.YearsExperience, data.Salary)
plt.show
```

```python
x = data.iloc[:, :-1].values  # Independent Variable, YearsExperience
y = data.iloc[:,-1].values  # Dependent Variable, Salary
```

```python
# Split the data into train and test

from sklearn.model_selection import train_test_split 

X_train, X_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=0)
```

```python
# Use your training data to train a linear regression model
from sklearn.linear_model import LinearRegression
linearModel = LinearRegression()

linearModel.fit(X_train, y_train)  # Train the model
```

```python
print(X_test)
print(y_test)
```

```python
y_pred = linearModel.predict(X_test)
y_pred
```

```python
plt.scatter(X_test, y_test, color="red")
plt.plot(X_test, linearModel.predict(X_test), color='green')
plt.show()
```
