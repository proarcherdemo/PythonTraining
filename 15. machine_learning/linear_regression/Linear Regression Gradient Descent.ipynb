```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("StudentsPerformance.csv")
df
```

```python
df.info()
```

```python
plt.scatter(df['math score'], df['reading score'])
plt.show()
```

```python
X = df[['math score']]   # independent variable
y = df['reading score']  # dependent variable
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)
```

```python
from sklearn.linear_model import LinearRegression, SGDRegressor
from sklearn.metrics import r2_score, mean_squared_error
from sklearn.preprocessing import StandardScaler

# Linear Regression - OLS

linearModel = LinearRegression()
linearModel.fit(X_train, y_train)  # Train the model
y_pred = linearModel.predict(X_test)
r2 = r2_score(y_test, y_pred)
mse = mean_squared_error(y_test, y_pred)

print(r2)
print(mse)
```

```python
sgdModel = SGDRegressor(max_iter=1000, learning_rate="constant", eta0=0.01, random_state=42)

sc = StandardScaler()
X_train_scaled = sc.fit_transform(X_train)
#y_train_scaled = sc.fit_transform(y_train)

print(X_train_scaled)

X_test_scaled = sc.fit_transform(X_test)
#y_test_scaled = sc.fit_transform(y_test)

sgdModel.fit(X_train_scaled, y_train)
y_pred_sgd= sgdModel.predict(X_test_scaled)

print(y_pred_sgd)
r2 = r2_score(y_test, y_pred_sgd)
mse = mean_squared_error(y_test, y_pred_sgd)

print(r2)
print(mse)
```
